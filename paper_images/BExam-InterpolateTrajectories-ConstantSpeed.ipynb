```python
def make_all_scatter_inds(samp_interval, maxval):
    all_scatter_inds = []
    temp = []
    next_to_add = 0
    while next_to_add < maxval:
        temp.append(next_to_add)
        all_scatter_inds.append(list(temp))
        next_to_add += samp_interval
    
    return all_scatter_inds
all_scatter_inds = make_all_scatter_inds(1000,10001)
```

```python
class LinearPiecewiseFunction(object):
    def __init__(self, xvals, yvals):
        self.xvals = xvals
        self.yvals = yvals
    
    def function(self,xval):
        low_ix = np.where(self.xvals <= xval)[0][-1]
        if low_ix == len(self.xvals)-1:
            return self.yvals[-1]
        xsep = self.xvals[low_ix+1]-self.xvals[low_ix]
        ysep = self.yvals[low_ix+1]-self.yvals[low_ix]
        
        return ((xval-self.xvals[low_ix])/xsep) * ysep + self.yvals[low_ix]
```

```python
retimed_trajs = []
for i in range(2):
    # just do first traj for now
    distance = []
    distance.append(0)
    traj = all_traj[i][0]
    for ix in range(1,len(traj)):
        distance.append(np.sqrt(np.sum(np.square(traj[ix] - traj[ix-1]))))
    distance = np.array(distance)
    newtimes = np.cumsum(distance)/np.sum(distance)
    original_times = np.linspace(0,1,len(newtimes))
    f = LinearPiecewiseFunction(newtimes, original_times)
    warped_time = np.array([f.function(ot) for ot in original_times])
    pos = LinearPiecewiseFunction(original_times, traj)
    new_traj = np.array([pos.function(wt) for wt in warped_time])
    for axis in range(1):
        plt.plot(new_traj[:,axis])
        plt.plot(traj[:,axis])
    new_traj = np.expand_dims(new_traj,0)
    retimed_trajs.append(new_traj)
```
